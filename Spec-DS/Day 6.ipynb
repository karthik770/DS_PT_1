```python
#series ->one dimensional array with indexed items (string , integer, float)
import pandas as pd
a=pd.Series(['a','b','c'])
print(a)
print(a.values)
print(a.index)
#Accessing pandas series obj by address
print(a[0])
print(a[0:1])
#Customizing the pandas series obj
b=pd.Series(['a','b','c'],index=['i','j','k'])
print(b)
#Acessing pandas series obj by data index
print(b['k'])
 
#create a float values and access the data using index and address
c = pd.Series([1.1,1.2,1.3],index=['p','q','r'])
print(c)
```

```python
#Dataframe 
a=pd.DataFrame([1,2,3,4,5],columns=['ID'])
print(a)

#creating a dataframe from Dictionary

index_dict={'s1':1,'s2':2,'s3':3,'s4':4,'s5':5}
indexdata=pd.Series(index_dict)
b=pd.DataFrame(indexdata,columns=['Name'])
print(b)
#Creating a pandas df using numpy 
import numpy as np
c=pd.DataFrame(np.random.rand(3,2),columns=['a','b'],index=['i','j','k'])
print(c)
d=pd.DataFrame(np.random.rand(3,3),columns=['a','b','c'],index=['i','j','k'])
print(d)
#create a pandas df using structured numpy array
A = np.zeros(3, dtype=[('A', 'i8'), ('B', 'f8')])
e=pd.DataFrame(A)
print(e)
```

```python
#Pandas Index Object -> It is immutable
indd=pd.Index([2,3,4,5,6,7])
print(indd)
print(indd[0])
#indd[1]=0 #-->Not work 

#Index object as set (intersection , union, symmetric diff)
ind1=pd.Index([2,3,4,5,6,7])
ind2=pd.Index([8,3,10,7,12,13])
print(pd.Index.intersection(ind1,ind2)) #intersection
print(pd.Index.union(ind1,ind2)) #Union
print(pd.Index.symmetric_difference(ind1,ind2)) #symmetric difference
```

```python
#Series
a=pd.Series([1,2,3,4,5,6],index=['i','j','k','l','m','n'])
print(a) # Print
print(a.keys()) # to get the index
print(list(a.items())) #ti get the list of value 
print(a['l']) # to get the value with index
print('z' in a) #to check index present here ot not
a['i']=10 # assigning of data
print(a) #print
b=a.loc[['i','n']]  # Getting a value using index by loc keyword
print(b)
c=a.iloc[2] #Getting a value using a iloc keyword
print(c)
```

```python
#Dataframe indexing
a=pd.DataFrame([1,2,3,4,5],columns=['ID'])
print(a.values) #Get the value
print(a.T) #Transpose 
print(a.values[2]) # access using index
```

```python
#Index Indexing / Selction
indd=pd.Index([2,3,4,5,6,7])
print(indd[indd>4]) # Adding logic to index 
print(indd[0]) # selecting from index
```

Index Preservation:
    Operating on data in pandas:
    Understanding how numpy is helping with Pandas DS(Series, DF, Index)
    

```python
import numpy as np
import pandas as pd
rng= np.random.RandomState(42)
ser=pd.Series(rng.randint(0,10,4)) # randint(start, end , countofno)
print(ser)
df = pd.DataFrame(rng.randint(0, 10, (3, 4)),
                  columns=['A', 'B', 'C', 'D'])
print(df)
#Universal function  of numpy 
# If u apply numpy ufunc to any of the pandas DS (series, dataframe), the result will be another pandas object with index preserved
ip=np.exp(ser)
print(ip)
ip2=np.sin(df*np.pi/4)
print(ip2)
```

Index Alignment: 
1.While performing Binary operation , if there is incomplete data in any one of the input, then index alignment will come..
2.Ex: Series , Data frame


```python
#Index alignment in Series:
a=pd.Series({'S1':70,'S2':90,'S3':100},name='Marklist1')
b=pd.Series({'S2':90,'C2':92,'C3':100},name='Marklist2')
cal=a+b
print(cal)
#Handling the missing value
a.add(b,fill_value=0)
```

```python
#Index alignment in DataFrame
r=np.random.RandomState(42)
a=pd.DataFrame(r.randint(0,20,(2,2)),columns=list('AB'))
print(a)
b=pd.DataFrame(r.randint(0,10,(3,3)),columns=list('ABC'))
print(b)
print(a+b)
#Handling the missing data
#fillcustom=a.stack().mean()
a.add(b,fill_value=0)
```
