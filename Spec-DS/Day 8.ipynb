Merge and Join Datasets	
    =>pd.merge
    Merge and Join
	Relational Algebra
    =>merge, join 
	Categories of Joins
	Specification of the Merge Key
	Specifying Set Arithmetic for Joins
	The suffixes Keyword

```python
#Merge
#The two df should have one same column or index 
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})
c=pd.merge(df1,df2)
print(c)
#Join
print(pd.merge(df1,df2,how='inner'))
print(pd.merge(df1,df2,how='outer'))
print(pd.merge(df1,df2,how='left'))
print(pd.merge(df1,df2,how='right'))
```

```python
#Merge keys:
#on
#left_on
#right_on
#ON keyword:
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})
df3 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'salary': [70000, 80000, 120000, 90000]})
print(pd.merge(df1,df2, on='employee'))
print(pd.merge(df1,df3, left_on='employee', right_on='name'))
```

```python
#Suffix keyyword
dfc1 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'rank': [1, 2, 3, 4]})
dfc2 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'rank': [3, 1, 4, 2]})
print(pd.merge(dfc1,dfc2,on='name' ,suffixes=["_c1", "_c2"]))
```

GroupBy: 
Split=>break the value and then group 
Apply=>apply agg, filtering , Tranformation 
Combine=>Merge


```python
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data': range(6)}, columns=['key', 'data'])
#print(df)
#Group by 
print(df.groupby('key').sum())
#Apply

b=df.groupby('key').aggregate(['min', np.median, max])
#print(b)
c=df.groupby('key').transform(lambda x: x +5)
#print(c)
c1=df.groupby('key').transform(lambda x:x*5)
c2=df.groupby('key').transform(lambda x:x//2)
#print(c1)
#print(c2)

#filter
d=df.filter('data') # Filtering a column name
print(d)

# Filering with logic

f1=['a','b','c','d','e']
def pickthechar(a):
    i=['a','e']
    if(a in i):
        return True
    else:
        return False
fil1=filter(pickthechar,f1)
print(fil1)
for j in fil1:
    print(j)
```

```python
#Combine=>Merge

df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'name': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})
print(pd.merge(df1,df2, left_on='employee', right_on='name'))
#Renaming
df2.rename(columns={'name':'employee'},inplace=True) #Rename the column 
print(df2.columns)
print(pd.merge(df1, df2, on='employee'))
```
