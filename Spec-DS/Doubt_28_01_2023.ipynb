Random : Computer will gen the random value , Numpy 


```python
import numpy as np
import pandas as pd
a=np.array([1,2,3,4,5])
print(a)
b=np.random.rand(3,2) #3 dim with 2 size
print(b)
c=np.random.randint(10,size=(4,3)) # setting the limit value to generate the random value
print(c)
rd=np.random.RandomState(42) 
d=pd.DataFrame(rd.randint(0,10,(4,3)))
print(d)
```
